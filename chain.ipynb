```python
from langchain.agents import load_tools, initialize_agent, AgentType
from langchain.llms.openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
def get_api_key():
    load_dotenv()
    # base_model = os.environ['BASE_MODEL']
    finetune_model = os.environ['FINETUNED_MODEL']
    openai_key = os.environ['OPENAI_API_KEY']
    return finetune_model, openai_key
```

```python
model_name, openai_api_key = get_api_key()
```

```python
llm = OpenAI(model=model_name, api_key=openai_api_key)
tools = load_tools(['wikipedia',  'terminal'], allow_dangerous_tools=True, llm=llm)

datascie = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```python
datascie.run("what is wikidata uri of dog")
```
